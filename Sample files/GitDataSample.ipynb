```python
import json
import pandas as pd

file_path = r'G:\BigData\data.json'

repo_records = []
commit_records = []
follower_records = []
following_records = []

with open(file_path, 'r', encoding='utf-8') as f:
    for i, line in enumerate(f):
        try:
            record = json.loads(line)
            user_login = record.get('login')
            user_id = record.get('id')

            # --- repo_list ---
            for repo in record.get('repo_list') or []:
                flat_repo = repo.copy()
                flat_repo['user_login'] = user_login
                flat_repo['user_id'] = user_id
                repo_records.append(flat_repo)

            # --- commit_list ---
            for commit in record.get('commit_list') or []:
                flat_commit = commit.copy()
                flat_commit['user_login'] = user_login
                flat_commit['user_id'] = user_id
                commit_records.append(flat_commit)

            # --- follower_list ---
            for follower in record.get('follower_list') or []:
                flat_follower = {
                    'follower_login': follower,
                    'user_login': user_login,
                    'user_id': user_id
                }
                follower_records.append(flat_follower)

            # --- following_list ---
            for following in record.get('following_list') or []:
                flat_following = {
                    'following_login': following,
                    'user_login': user_login,
                    'user_id': user_id
                }
                following_records.append(flat_following)

        except Exception as e:
            print(f"Error on line {i+1}: {e}")

        if i >= 500:  # 🔁 Adjust this to load more/less users
            break

# Convert each into DataFrames
repo_df = pd.DataFrame(repo_records)
commit_df = pd.DataFrame(commit_records)
follower_df = pd.DataFrame(follower_records)
following_df = pd.DataFrame(following_records)

# Preview results
print("📦 Repo DF:")
display(repo_df.head())

print("🧾 Commit DF:")
display(commit_df.head())

print("👥 Follower DF:")
display(follower_df.head())

print("➡️ Following DF:")
display(following_df.head())
```
